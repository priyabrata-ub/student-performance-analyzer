```python
import pandas as pd
import numpy as np
from faker import Faker
import random

print("Everything installed successfully!")
```

```python
fake = Faker('en_IN')
```

```python
casts = [
    "General",
    "OBC",
    "SC",
    "ST",
    "EWS"
]

subjects = [
    "English",
    "Maths",
    "Biology",
    "Chemistry",
    "Physics",
    "Computer Applications"
]
```

```python
students = []
```

```python
for i in range(1, 101):
    
    student = {

        "Roll_Number": f"2026CS{i:03}",

        "Student_Name": fake.name(),

        "English": random.randint(35, 100),

        "Maths": random.randint(35, 100),

        "Biology": random.randint(35, 100),

        "Chemistry": random.randint(35, 100),

        "Physics": random.randint(35, 100),

        "Computer_Applications": random.randint(35, 100),

        "Attendance": random.randint(50, 100),

        "Date_of_Birth": fake.date_of_birth(
            minimum_age=17,
            maximum_age=22
        ),

        "Phone_Number": fake.phone_number(),

        "Caste": random.choice(casts),

        "Father_Name": fake.name_male(),

        "Mother_Name": fake.name_female(),

        "Family_Income_Per_Annum":
            random.randint(100000, 2000000),

        "Address":
            fake.address().replace("\n", ", ")
    }

    students.append(student)
```

```python
df = pd.DataFrame(students)
```

```python
import pandas as pd

# Load student dataset
students = pd.read_csv("academic_dataset.csv")

# Load scholarship dataset
scholarships = pd.read_csv("scholarship_dataset.csv")

# Select first student
for i in range(len(students)):
    
    student = students.iloc[i]

    total_marks = sum(student[sub] for sub in subjects)
    percentage = total_marks / len(subjects)

    if student["Family_Income_Per_Annum"] < 500000:
        print(
            student["Student_Name"],
            percentage,
            student["Family_Income_Per_Annum"],
            student["Caste"]
        )
        break

# Calculate percentage
subjects = [
    "English",
    "Maths",
    "Biology",
    "Chemistry",
    "Physics",
    "Computer_Applications"
]

total_marks = sum(student[sub] for sub in subjects)
percentage = total_marks / len(subjects)

print("Student Name:", student["Student_Name"])
print("Percentage:", round(percentage, 2))
print("Income:", student["Family_Income_Per_Annum"])
print("Caste:", student["Caste"])

print("\nEligible Scholarships:\n")

found = False

for _, scholarship in scholarships.iterrows():

    if (
        percentage >= scholarship["Min_Percentage"]
        and student["Family_Income_Per_Annum"] <= scholarship["Max_Income"]
        and (
            scholarship["Caste"] == "Any"
            or scholarship["Caste"] == student["Caste"]
        )
    ):
        print("✓", scholarship["Scholarship_Name"])
        found = True

if not found:
    print("❌ No Scholarship Eligible")
```
